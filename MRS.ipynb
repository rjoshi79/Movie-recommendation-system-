```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credit=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
movies.info()
movies.describe()
```

```python
credit.head()
```

```python
movies = movies.merge(credit,on='title')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
# For convert string of List to normal list
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
# literal_eval convert string list in to normal list

        
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
def fetch_job(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_job)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#  For Remove Space between two words
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# concate all columns in tags column 
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

movies.head()
```

```python
# create new df 
new_movies=movies[['movie_id','title','tags']]
new_movies.head()
```

```python
# convert list to string
new_movies['tags'] = new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies.head()
```

```python
# convert into lower case
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# for words stemelization
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
new_movies.head()
```

```python
# for convert words to vectors for perfect recommendation

from sklearn.feature_extraction.text  import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out()
```

```python
# this function can give similarity between two vectors 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vector)
```

# Main Function

```python
def Recommend(movie):
    movie_index=new_movies[new_movies['title'] == movie].index[0]
    distance = similar[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
Recommend('Batman Begins')
```
